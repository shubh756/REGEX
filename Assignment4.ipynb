**Numeric-Numeric Analysis**
Analyzing the two numeric variables from a dataset is known as numeric-numeric analysis. We can analyze it in three different ways.
- Scatterplot
- Pairplot
- Barplot
- Piechart
- Lineplot

**Scatterplot**
- The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

```python
plt.figure(figsize = (30, 15))
plt.xticks(rotation = 90)
sns.scatterplot(x = 'region', y = 'vodka', hue = 'year', data = df)
```

```python
plt.figure(figsize = (20, 20))
plt.xticks(rotation = 90)
sns.scatterplot(x = 'region', y = 'beer', hue = 'year', data = df)
```

**Pairplot**
- A pairplot plot a pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. 

```python
sns.pairplot(data = df , vars = ['wine', 'vodka', 'brandy', 'champagne', 'beer'], hue = 'year')
```

**Barplot**
- Barplot will show the relation between two numeric data

```python
plt.figure(figsize = (15,10))
sns.barplot(df['year'],df['vodka'])
```

```python
plt.figure(figsize = (15,10))
sns.barplot(df['year'],df['champagne'])
```

**Piechart**
- A Pie Chart is a circular statistical plot that can display only one series of data.

```python
plt.figure(figsize = (12,12))
x = [df['wine'].mean(), df['beer'].mean(),df['vodka'].mean(),df['champagne'].mean(),df['brandy'].mean()]
alcohol = ['Wine', 'Beer', 'Vodka', 'Champagne', 'Brandy']
plt.title('Mean of various alchols' , fontsize = 20)
plt.pie(x, labels = alcohol)
```

**Lineplot**
- The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters. These parameters control what visual semantics are used to identify the different subsets.

```python
plt.figure(figsize = (10, 5))
plt.style.use('seaborn-darkgrid')
sns.lineplot(data = df, x = 'year', y = 'wine', label = 'Wine')
sns.lineplot(data = df, x = 'year', y = 'beer', label = 'Beer')
sns.lineplot(data = df, x = 'year', y = 'vodka', label = 'Vodka')
sns.lineplot(data = df, x = 'year', y = 'champagne', label = 'Champagne')
sns.lineplot(data = df, x = 'year', y = 'brandy', label = 'Brandy')
```
